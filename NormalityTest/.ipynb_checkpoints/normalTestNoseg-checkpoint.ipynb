```python
import numpy as np
import scipy as sp
import pingouin as pg
```

# NinaPro DB6

### First feature set: MAV

```python
auxName='mavMatrix'
carpet='ExtractedDataNinaDB6'
```

The Ninapro DB6 data set consists of the following columns: 14 Chs, person, day, time (1 = morning and 2 = afternoon), class, and number of repetitions.

```python
DataNinaproDB6=np.genfromtxt('../ExtractedData/'+carpet+'/'+auxName+'NoSeg.csv',delimiter=',')
print(DataNinaproDB6)
```

We use two multivariate normality tests: Henze-Zirkler test (H-Zt), and Shapiro-Wilk test (S-Wt).


```python
cont = 0
cont1 = 0
cont2 = 0
classes=[1,3,4,6,9,10,11]
for person in range(1,11):
    for day in range(1,6):
        for time in range(1,3):
            for cl in classes:
                aux=DataNinaproDB6[np.where((DataNinaproDB6[:,14]==person)*(DataNinaproDB6[:,15]==day)*(DataNinaproDB6[:,16]==time)*(DataNinaproDB6[:,17]==cl)),0:14][0]
                if sp.size(aux)==0:
                    print("NO DATA. person=",person, "; day=" , day , "; time=" , time , "; class=" , cl)
                else:
                    normal, p = pg.multivariate_normality(aux, alpha=.05)
                    p1=sp.stats.shapiro(aux)
                    if normal == True:
                        cont1 += 1
              
                    if p1[1] <  0.05:
                        cont2 += 1
                        cont += 1
                    print("H-Zt: ",normal,p,"S-Wt:",p1,"person=",person, "; day=" , day , "; time=" , time , "; class=" , cl)
print ("# True H-Zt:" , cont1 ,"# True S-Wt:" , cont2,"Total:" , cont,cont1/cont,cont2/cont)
```

# Collected Data

### First feature set: MAV

```python
auxName='mavMatrix'
carpet='ExtractedDataCollectedData'
```

The Collected data set consists of the following columns: 8 Chs, type of data(0=training and 1=testing ), person, class, and number of repetitions.

```python
CollectedData=np.genfromtxt('../ExtractedData/'+carpet+'/'+auxName+'NoSeg.csv',delimiter=',')
print(CollectedData)
```

We use two multivariate normality tests: Henze-Zirkler test (H-Zt), and Shapiro-Wilk test (S-Wt).


```python
cont = 0
cont1 = 0
cont2 = 0

classes=[1,2,3,4,5]
for ty in range(0,2):
    for person in range(1,61):
        for cl in classes:
            aux=CollectedData[np.where((CollectedData[:,8]==ty)*(CollectedData[:,9]==person)*(CollectedData[:,10]==cl)),0:8][0]
            if sp.size(aux)==0:
                print("NO DATA. type=",ty, "person=",person, "class=" , cl)
            else:
                normal, p = pg.multivariate_normality(aux, alpha=.05)
                p1=sp.stats.shapiro(aux)
                if normal == True:
                    cont1 += 1
              
                if p1[1] <  0.05:
                    cont2 += 1
                cont += 1
                
                print("H-Zt: ",normal,p,"S-Wt:",p1,"; type=",ty, "; person=",person, "; class=" , cl)
print ("# True H-Zt:" , cont1 ,"# True S-Wt:" , cont2,"Total:" ,cont1/cont,cont2/cont)
```
